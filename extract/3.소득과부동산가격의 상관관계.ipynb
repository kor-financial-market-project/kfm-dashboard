```python
import pandas as pd
```

```python
realstate = pd.read_csv("../data/isocode_added_salesprice.csv")
grdp = pd.read_csv("../data/시도별_1인당_지역내총생산__지역총소득__개인소득_20240516162050.csv")
```

```python
grdp_df = grdp.drop(0)
```

```python
grdp_df # 1인당 지역내총생산,	1인당 지역총소득,	1인당 개인소득,	1인당 민간소비
```

```python
# 데이터프레임을 melt하여 long-form으로 변환
df_melted = grdp_df.melt(id_vars=["시도별"], var_name="연도", value_name="값")
```

```python
df_melted.loc[df_melted['시도별']=='전국']
```

```python
# 날짜에서 연도만 추출
df_melted["연도"] = df_melted["연도"].str.extract(r'(\d{4})')
```

```python
df_melted.loc[df_melted['시도별']=='전국']
```

```python
df_melted['순서'] = df_melted.groupby(['시도별', '연도']).cumcount()
value_names = ['1인당 지역내총생산', '1인당 지역총소득', '1인당 개인소득', '1인당 민간소비']
df_melted['값 이름'] = df_melted['순서'].apply(lambda x: value_names[x])
```

```python
df_melted.sort_values(['시도별','연도'])
```

```python
# pivot_table을 사용하여 원하는 형태로 데이터 변환
df_pivot = df_melted.pivot_table(index=['시도별', '연도'], columns='값 이름', values='값', aggfunc='first').reset_index().rename_axis(None,axis=1)
```
