```python
import requests

api_key = 'api_key 발급 후 입력'
url = f'http://ecos.bok.or.kr/api/StatisticSearch/{api_key}/json/kr/1/10000/513Y001/M/201905/202405/?/?/?/?' # 통계 코드 및 항목 코드 참고

response = requests.get(url)
data = response.json()

print(data)
```

```python
import pandas as pd

# 데이터 추출
items = data['StatisticSearch']['row']

# DataFrame 생성
df = pd.DataFrame(items)

df
```

```python
# 데이터 전처리

df = df.drop(columns=['ITEM_CODE2', 'ITEM_NAME2', 'ITEM_CODE3', 'ITEM_NAME3', 'ITEM_CODE4', 'ITEM_NAME4', 'WGT'])

df
```

```python
# CSV 파일로 저장
csv_file_path = '/content/drive/My Drive/ESI_201905_202405.csv'
df.to_csv(csv_file_path, index=False)
```
